## Step 8: Analyze Video Duration Patterns (with Pure Python ISO 8601 Conversion)

Analyze and visualize trending video durations using a pure Python function to convert ISO 8601 durations (e.g., PT15M10S) to seconds.


```python
import re

def parse_duration(duration):
    """
    Parse an ISO 8601 duration string (e.g., 'PT2H15M10S') into seconds.
    Works for durations containing H, M, and S (e.g., 'PT10M15S').
    """
    if not isinstance(duration, str) or not duration.startswith('PT'):
        return None

    hours = minutes = seconds = 0
    match_hours = re.search(r'(\d+)H', duration)
    match_minutes = re.search(r'(\d+)M', duration)
    match_seconds = re.search(r'(\d+)S', duration)
    if match_hours:
        hours = int(match_hours.group(1))
    if match_minutes:
        minutes = int(match_minutes.group(1))
    if match_seconds:
        seconds = int(match_seconds.group(1))
    total_seconds = hours * 3600 + minutes * 60 + seconds
    return total_seconds

df_focus['video_duration_sec'] = df_focus['video_duration'].apply(parse_duration)

# Plot distribution of durations (all trending videos)
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.histplot(df_focus['video_duration_sec'].dropna(), bins=100, kde=True)
plt.title('Distribution of Trending Video Durations (Seconds)')
plt.xlabel('Duration (seconds)')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('trending_video_duration_distribution.png')
plt.show()

# Boxplot of durations by top 5 categories
top5_cats = df_focus['video_category_id'].value_counts().index[:5]
plt.figure(figsize=(10,5))
sns.boxplot(
    data=df_focus[df_focus['video_category_id'].isin(top5_cats)],
    x='video_category_id',
    y='video_duration_sec'
)
plt.title('Video Duration by Top 5 Categories')
plt.xlabel('Category ID')
plt.ylabel('Duration (seconds)')
plt.tight_layout()
plt.savefig('duration_by_top5_categories.png')
plt.show()

# Boxplot of durations by top 5 countries
top5_countries = df_focus['video_trending_country'].value_counts().index[:5]
plt.figure(figsize=(10,5))
sns.boxplot(
    data=df_focus[df_focus['video_trending_country'].isin(top5_countries)],
    x='video_trending_country',
    y='video_duration_sec'
)
plt.title('Video Duration by Top 5 Countries')
plt.xlabel('Country')
plt.ylabel('Duration (seconds)')
plt.tight_layout()
plt.savefig('duration_by_top5_countries.png')
plt.show()
```

## Step 9: Analyze Engagement Metrics

- Compute correlations between views, likes, and comments.
- Visualize the relationships using scatterplots, colored by category.
- Save each figure for use in reports or a portfolio.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Correlation matrix
corr = df_focus[['video_view_count', 'video_like_count', 'video_comment_count']].corr()
print("Correlation Matrix:\n", corr)

# Scatterplot: Views vs Likes
plt.figure(figsize=(8,6))
sns.scatterplot(
    x='video_view_count', y='video_like_count', 
    data=df_focus.sample(5000, random_state=42), alpha=0.3
)
plt.title('Views vs Likes (Sample of 5000)')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.tight_layout()
plt.savefig('views_vs_likes.png')
plt.show()

# Scatterplot: Views vs Comments
plt.figure(figsize=(8,6))
sns.scatterplot(
    x='video_view_count', y='video_comment_count', 
    data=df_focus.sample(5000, random_state=1), alpha=0.3
)
plt.title('Views vs Comments (Sample of 5000)')
plt.xlabel('Views')
plt.ylabel('Comments')
plt.tight_layout()
plt.savefig('views_vs_comments.png')
plt.show()
```

## Step 10: Engagement Metrics by Category

Analyze average views, likes, and comments per trending video for each of the top 10 categories. This reveals which content types generate the most engagement—not just which trend most often.


```python
# Group by category and calculate mean engagement metrics
cat_engagement = df_focus.groupby('video_category_id')[['video_view_count', 'video_like_count', 'video_comment_count']].mean().sort_values('video_view_count', ascending=False).head(10)

print(cat_engagement)

# Plot
cat_engagement.plot(kind='bar', figsize=(12,6))
plt.title('Average Engagement by Video Category (Top 10)')
plt.ylabel('Average Count')
plt.xlabel('Category ID')
plt.tight_layout()
plt.savefig('avg_engagement_by_category.png')
plt.show()
```

## Step 11: Engagement Metrics by Country

Analyze average views, likes, and comments per trending video for the top 10 countries. This reveals which regions drive the highest audience engagement.


```python
# Group by country and calculate mean engagement metrics
country_engagement = df_focus.groupby('video_trending_country')[['video_view_count', 'video_like_count', 'video_comment_count']].mean().sort_values('video_view_count', ascending=False).head(10)

print(country_engagement)

# Plot
country_engagement.plot(kind='bar', figsize=(12,6))
plt.title('Average Engagement by Country (Top 10)')
plt.ylabel('Average Count')
plt.xlabel('Country')
plt.tight_layout()
plt.savefig('avg_engagement_by_country.png')
plt.show()
```
